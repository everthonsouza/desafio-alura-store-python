### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
import matplotlib.pyplot as plt

def faturamento(loja, nome_loja):
  fat = round(loja['Preço'].sum(), 2)
  print(f'A loja "{nome_loja}" faturou R${fat}')
  return fat
```

```python
fat1 = faturamento(loja, 'loja 1')
fat2 = faturamento(loja2, 'loja 2')
fat3 = faturamento(loja3, 'loja 3')
fat4 = faturamento(loja4, 'loja 4')

nomes = ['loja 1', 'loja 2', 'loja 3', 'loja 4']
faturamentos = [fat1, fat2, fat3, fat4]

plt.bar(nomes, faturamentos)
plt.title("Faturamento Total por Loja")
plt.ylabel("Montante em Vendas (R$)")
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
def mostrar_valores_reais(pct, all_vals):
    valor_absoluto = int(round(pct/100. * sum(all_vals)))
    return f'{valor_absoluto}'

categoria_1 = loja['Categoria do Produto'].value_counts()
plt.pie(categoria_1, labels=loja['Categoria do Produto'].unique(), autopct=lambda x: mostrar_valores_reais(x, categoria_1))
plt.title('Vendas por Categoria - Loja 1')
plt.show()
```

```python
categoria_2 = loja2['Categoria do Produto'].value_counts()
plt.pie(categoria_2, labels=loja['Categoria do Produto'].unique(), autopct=lambda x: mostrar_valores_reais(x, categoria_1))
plt.title('Vendas por Categoria - Loja 2')
plt.show()
```

```python
categoria_3 = loja3['Categoria do Produto'].value_counts()
plt.pie(categoria_3, labels=loja['Categoria do Produto'].unique(), autopct=lambda x: mostrar_valores_reais(x, categoria_1))
plt.title('Vendas por Categoria - Loja 3')
plt.show()
```

```python
categoria_4 = loja4['Categoria do Produto'].value_counts()
plt.pie(categoria_4, labels=loja4['Categoria do Produto'].unique(), autopct=lambda x: mostrar_valores_reais(x, categoria_1))
plt.title('Vendas por Categoria - Loja 4')
plt.show()
```

```python
categoria_1 = loja.groupby('Categoria do Produto').sum()['Preço']
categoria_2 = loja2.groupby('Categoria do Produto').sum()['Preço']
categoria_3 = loja3.groupby('Categoria do Produto').sum()['Preço']
categoria_4 = loja4.groupby('Categoria do Produto').sum()['Preço']

plt.figure(figsize=(12, 5))

plt.plot(categoria_1, marker='o')
plt.plot(categoria_2, marker='o')
plt.plot(categoria_3, marker='o')
plt.plot(categoria_4, marker='o')
plt.xlabel('Categoria do Produto')
plt.ylabel('Faturamento')
plt.title('Faturamento por Categoria')
plt.legend(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'])
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_1 = round(float(loja['Avaliação da compra'].sum() / len(loja)), 2)
media_2 = round(float(loja2['Avaliação da compra'].sum() / len(loja2)), 2)
media_3 = round(float(loja3['Avaliação da compra'].sum() / len(loja3)), 2)
media_4 = round(float(loja4['Avaliação da compra'].sum() / len(loja4)), 2)

print(f'A loja 1 teve uma média de {media_1} estrelas')
print(f'A loja 2 teve uma média de {media_2} estrelas')
print(f'A loja 3 teve uma média de {media_3} estrelas')
print(f'A loja 4 teve uma média de {media_4} estrelas')

plt.barh(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'], [media_1, media_2, media_3, media_4])
plt.xlabel('Média de Avaliação')
plt.ylabel('Lojas')
plt.title('Média de Avaliação por Loja')
for i, valor in enumerate([media_1, media_2, media_3, media_4]):
  plt.text(valor, i, f'{valor}', va='center', ha='left')

plt.show()
```

```python
lista_media_categorias = {}

def calcula_media_loja_categoria(nome_loja, loja):
  media = loja.groupby('Categoria do Produto').sum()['Avaliação da compra']

  for categoria in loja['Categoria do Produto'].unique():
    lista_media_categorias[f'{categoria}_{nome_loja}'] = round(float(media[categoria] / len(loja[loja['Categoria do Produto'] == categoria])), 2)

calcula_media_loja_categoria('loja 1', loja)
calcula_media_loja_categoria('loja 2', loja2)
calcula_media_loja_categoria('loja 3', loja3)
calcula_media_loja_categoria('loja 4', loja4)

lista_media_categorias = dict(sorted(lista_media_categorias.items(), key=lambda item: item[1]))

plt.figure(figsize=(14, 6))

plt.bar(lista_media_categorias.keys(), lista_media_categorias.values())
plt.xlabel('Categorias_Loja')
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Categoria e Loja')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
def montar_lista_produtos(loja):
  lista_vendas = {}
  lista_produtos = loja['Produto'].unique()
  for produto in lista_produtos:
    lista_vendas[produto] = int(loja.groupby('Produto').count()['Preço'][produto])
  return dict(sorted(lista_vendas.items(), key=lambda item: item[1])) #retorna a lista ordenada

def mostrar_vendas(lista, nome_loja):
  print(f'{nome_loja}: Produtos mais vendidos: "{list(lista.keys())[-1]}" e "{list(lista.keys())[-2]}"')
  print(f'{nome_loja}: Produtos menos vendidos: "{list(lista.keys())[0]}" e "{list(lista.keys())[1]}"')

lista_loja_1 = montar_lista_produtos(loja)
lista_loja_2 = montar_lista_produtos(loja2)
lista_loja_3 = montar_lista_produtos(loja3)
lista_loja_4 = montar_lista_produtos(loja4)

mostrar_vendas(lista_loja_1, 'Loja 1')
mostrar_vendas(lista_loja_2, 'Loja 2')
mostrar_vendas(lista_loja_3, 'Loja 3')
mostrar_vendas(lista_loja_4, 'Loja 4')
```

# 5. Frete Médio por Loja


```python
frete_loja_1 = round(loja['Frete'].mean(), 2)
frete_loja_2 = round(loja2['Frete'].mean(), 2)
frete_loja_3 = round(loja3['Frete'].mean(), 2)
frete_loja_4 = round(loja4['Frete'].mean(), 2)

print(f'A loja 1 teve o custo médio de R$ {frete_loja_1} em fretes.')
print(f'A loja 2 teve o custo médio de R$ {frete_loja_2} em fretes.')
print(f'A loja 3 teve o custo médio de R$ {frete_loja_3} em fretes.')
print(f'A loja 4 teve o custo médio de R$ {frete_loja_4} em fretes.')
```

```python
plt.barh(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'], [frete_loja_1, frete_loja_2, frete_loja_3, frete_loja_4], color=['red', 'blue', 'green', 'yellow'])
plt.xlabel('Valor médio do Frete')
plt.ylabel('Lojas')
plt.title('Média de Custo com Frete')
for i, valor in enumerate([frete_loja_1, frete_loja_2, frete_loja_3, frete_loja_4]):
  plt.text(valor, i, f'R${valor}', va='center', ha='left')

plt.show()
```

```python
lucro_loja_1 = float(loja['Preço'].sum() - loja['Frete'].sum())
lucro_loja_2 = float(loja2['Preço'].sum() - loja2['Frete'].sum())
lucro_loja_3 = float(loja3['Preço'].sum() - loja3['Frete'].sum())
lucro_loja_4 = float(loja4['Preço'].sum() - loja4['Frete'].sum())

plt.barh(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'], [lucro_loja_1, lucro_loja_2, lucro_loja_3, lucro_loja_4])
for i, valor in enumerate([lucro_loja_1, lucro_loja_2, lucro_loja_3, lucro_loja_4]):
  plt.text(valor, i, f'R${round(valor, 2)}', va='center', ha='left')
plt.title('Faturamento Liquido por Loja (Fat - Custo Frete)')
plt.xlabel('Loja')
plt.show()
```

# Relatorio Final

**1 - Faturamento**

Analisando o faturamento das quatro lojas, a Loja 1 possui o maior faturamento e a Loja 4 o faturamento mais baixo, faturando respectivamente 1.534.509,12 contra 1.384.497,58 reais.


**2 - Vendas por Categoria**

As categorias mais vendias das lojas são:
 - Loja 1 - Móveis e Eletrônicos
 - Loja 2 - Móveis e Eletrônicos
 - Loja 3 - Móveis e Eletrônicos
 - Loja 4 - Móveis e Eletrodomésticos

Agora analisando o faturamento por categorias, vimos que o volume de vendas demonstra uma visão diferente, pois o volume de vendas de determinada categoria não tem um impacto tão direto quando olhamos o faturamento daquela categoria, então as categorias "Eletrônicos" e "Eletrodomésticos" são as categorias com maior faturamento em ambas lojas respectivamente.


**3 - Avaliações das Lojas**

 - A loja 1 teve uma média de 3.98 estrelas
 - A loja 2 teve uma média de 4.04 estrelas
 - A loja 3 teve uma média de 4.05 estrelas
 - A loja 4 teve uma média de 4.0 estrelas

Sendo assim, mesma a Loja 1 sendo a com maior faturamento, possui uma média de avaliações menor dentre as demais, vale realçar, que a diferença dessa média é bem pequena comparando as quatro lojas.

Quebrando essas avalições por categoria de produtos, vemos que a Loja 1 possui menores avaliações nas vendas de "Utilidades Domésticas" e "Esporte e Lazer" e a Loja 4 possui menor satisfação na venda de "Instrumentos Musicais" e "Eletrodomésticos".


**4 - Produtos Mais e Menos Vendidos**

Segue a relação dos produtos mais e menos vendidos:
 - Loja 1: Produtos mais vendidos: "Guarda roupas" e "TV Led UHD 4K"
 - Loja 1: Produtos menos vendidos: "Headset" e "Celular ABXY"

 - Loja 3: Produtos mais vendidos: "Kit banquetas" e "Cama king"
 - Loja 3: Produtos menos vendidos: "Blocos de montar" e "Jogo de copos"
 - Loja 4: Produtos mais vendidos: "Cama box" e "Faqueiro"

Analisando os produtos mais vendidos, vemos o impacto direto no volume de vendas das lojas e também no faturamento, sendo produtos das categorias "Móveis" e "Eletrodomésticos", essa análise pode mostrar para as lojas, quais os produtos e categorias que elas podem investir mais, tanto em propaganda quanto em melhor atendimento, buscando alanvar as vendas e com isso melhores resultados.


**5 - Frete por Loja**

Abaixo a relação do custo médio de cada loja:

A loja 1 teve um custo maior com frete e a loja 4 teve um custo menor, lembrando do faturamento e do volume de vendas, é previsivel que a loja que vende mais tem um custo maior de frete, sendo que o frete também tem relação direta com as categorias vendidas.

Ainda falando sobre o custo de frete, analisei que se subtrairmos o frete do faturamento, o resultado do ranking das lojas se mantem, ou seja, ainda sim a Loja 1 possui melhor resultado liquido e a Loja 4 possui o menor resultado.

A análise geográfica foi feita olhando as latitudes e longitudes das entregas e também a distribuição de vendas por estado.

Com base no gráfico de dispersão, vimos que as ambas lojas atendem regiões simulares, com exceção da Loja 4, que possui menor (em algumas regiões não possui) incidência de entregas.

Olhando por estados, vimos que ambas as lojas atenderam ambos estados, mas a loja 4 possui menor indicência em diversos estados.


**Conclusão**

Com base nos dados, visões e gráficos, minha sugestão é que a Loja 4 seja a loja a ser fechada, seguem as justificativas:
 - Ela possui o menor faturamento
 - Por mais que a categoria "Eletrodomésticos" esteja no ranking das categorias mais vendidas, essa também está bem pontuada nas demais lojas
 - A loja 4 não possui a maior avalição entre as lojas mas também não tem a melhor, sendo a diferença entre as lojas pequena
 - O produto mais vendido nessa loja está dentro das categorias mais vendidas nas demais lojas
 - O custo do frete é o menor, mas não tem um impacto tão grande ao ponto de destacar o resultado da loja, sendo ainda a com pior resultado liquido
 - A loja 4 foi a unica que não entregou em determinadas regiões
 - Todas as regiões quais a loja 4 se destaca, também são atendidas com qualidade pelas demais lojas
